```python
plt.boxplot(wage_data,vert=False)
plt.show()
```

```python
# step-1: cal q1
# step-2: cal q2
# step-3: cal q3
# step-4: iqr=q3-q1
# step-5: lb=q1-1.5*iqr
# step-6: ub=q3+1.5*iqr
# step-7: con1
# step-8: con2
# step-9: con
# step-10: gte the ocunt

q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
con1=wage_data<lb
con2=wage_data>ub
con3=con1|con2
outliers_df=visa_df[con3]
outliers_df['prevailing_wage']
outliers_df['prevailing_wage'].values
outliers_list=outliers_df['prevailing_wage'].values.tolist()
```

```python
outliers_list
```

```python
# i want to tell each observation of wage data
# if that observtaion will available in this 427 values
# then that observation will replace by mention
# otherwise that observation will be available as it is 

wage_medain=visa_df['prevailing_wage'].median()
wage_data=visa_df['prevailing_wage']
new_wage_data=[]
for i in wage_data:
    if i in outliers_list:
        new_wage_data.append(wage_medain)
    else:
        new_wage_data.append(i)
```

```python
plt.Figure(figsize=(10,6))
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(new_wage_data)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(new_wage_data,vert=False)
plt.show()
```

**winsorization**

- whatever the outliers less than lb fill with lb

- whatever the outliers greater than ub fill with ub

- in numpy we have method **clip**

```python
# step1: get the wage data values
# step2: cal lb
# step3: cal ub
# step4: check the parameters and apply in numpy np.clip()
# step5: again implement above visualization analysis

q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
np.clip(wage_data,lb,ub)
```
