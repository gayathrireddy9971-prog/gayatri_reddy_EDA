```python
#step-1: select the continent data 
#step-2: continet == 'Asia'   it is con1
#step-3: select the case data
#step-4: case status data == 'Certified'  is con2
#step-5: con=con1&con2
#step-6: apply main df on con get the len

visa_df['continent'].value_counts()
visa_df['case_status'].value_counts()

con1= visa_df['continent'] == 'Asia'
con2= visa_df['case_status'] == 'Certified'
con= con1&con2
len(visa_df[con])
```

```python
# from diff continent how many ppl are certified
labels=visa_df['continent'].unique()
cert,den=[],[]
for i in labels:
    con1= visa_df['continent'] == i
    con2= visa_df['case_status'] == 'Certified'
    con3= visa_df['case_status'] == 'Denied'
    cert_con= con1 & con2
    den_con= con1 & con3
    cert.append(len(visa_df[cert_con]))
    den.append(len(visa_df[den_con]))
    
cols=sorted(visa_df['case_status'].unique())
print(cols)

pd.DataFrame(zip(cert,den),
            columns=cols,index=labels)
```
