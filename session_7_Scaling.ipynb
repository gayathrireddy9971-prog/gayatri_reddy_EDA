**Zscore**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path=r'C:\Users\gayat\Documents\NARESH IT\Data Files\Visadataset.csv'
visa_df=pd.read_csv(file_path)

cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

```python
#step-1: take the pwage data
#step-2: calculate mean of data
#step-3: calculate sd of data
#step-4: Nr= data-mean step1-step2
#step-5: z wage=Nr/std = step4/step3
#step-6: create a new column visa_df['z_wage']=z_wage


o=visa_df['prevailing_wage']
n=visa_df['prevailing_wage'].mean()  #n=o.mean()
e=visa_df['prevailing_wage'].std()   #e=o.std()
Nr=o-n
z_wage=Nr/e
visa_df['z_wage']=z_wage
```

```python
visa_df[['prevailing_wage','z_wage']]
```

```python
plt.figure(figsize=(8,4))
plt.suptitle('Raw-data vs Z-data')
plt.subplot(2,2,1).hist(o)
plt.title('prevailing_wage')
plt.subplot(2,2,2).hist(z_wage)
plt.title('Z score prevailing_wage')
plt.subplot(2,2,3).boxplot(o,vert=False)
plt.subplot(2,2,4).boxplot(z_wage,vert=False)
plt.show()
```
