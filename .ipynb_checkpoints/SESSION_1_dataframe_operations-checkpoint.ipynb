**inplace**

- when w edo the operation the result output is not overwriting

- means the original dataframe is asame as it is

- in order to do this we need to use **inplace**

- by default inplace=False

- make **inplace=True** if you want to write the result

```python
df.drop('Marks',axis=1,inplace=True)
```

```python
df
```
